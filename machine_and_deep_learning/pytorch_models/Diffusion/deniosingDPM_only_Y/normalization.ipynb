```python
import numpy as np
from tqdm import tqdm
import os

# Directory containing the images
train_dir = os.path.join('resources', 'datasets', 'concatenated_outputs', 'train')
test_dir = os.path.join('resources', 'datasets', 'concatenated_outputs', 'test')

squares_Y, squares_Cb, squares_Cr = 0,0,0
count_Y, count_Cb, count_Cr = 0,0,0

def process_batch(batch):
    global squares_Y, squares_Cb, squares_Cr, count_Y, count_Cb, count_Cr, train_meanY, train_meanCb, train_meanCr

    # Separate components
    Y = batch[:64, :, :]
    Cb = batch[64:128, :, :]
    Cr = batch[128:, :, :]


    # Update min and max for each component
    squares_Y += np.sum((Y - train_meanY) **2)
    count_Y += len(Y.flatten())

    squares_Cb += np.sum((Cb - train_meanCb) **2)
    count_Cb += len(Cb.flatten())

    squares_Cr += np.sum((Cr - train_meanCr) **2)
    count_Cr += len(Cr.flatten())



# Process each image file
for filename in tqdm(os.listdir(train_dir)):
    filepath = os.path.join(train_dir, filename)
    
    # Load image with allow_pickle=True
    try:
        image = np.load(filepath, allow_pickle=True)
        process_batch(image)
    except:
        continue


train_std_Y = np.sqrt(squares_Y/count_Y)
train_std_Cr = np.sqrt(squares_Cr/count_Cr)
train_std_Cb = np.sqrt(squares_Cb/count_Cb)


print(train_std_Y)
print(train_std_Cb)
print(train_std_Cr)
```
